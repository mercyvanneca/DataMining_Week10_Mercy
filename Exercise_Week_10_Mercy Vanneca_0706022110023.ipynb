## Apriori Algorithm

We will use appriori algorithm to determine the frequently purchased products. 
For this case study, we will min_support=0.2

```python
frequent_itemsets = apriori(new_df, min_support=0.2, use_colnames=True)

# Print frequent itemsets
print("\nFrequent Itemsets:")
frequent_itemsets
```

Then, we will generate association rule of the frequent itemset based on confidence level with the threshold=0.6

```python
from mlxtend.frequent_patterns import association_rules

# Assuming you have already obtained frequent_itemsets using apriori

# Generate association rules with a minimum confidence of 0.6
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

# Print the generated rules
print("\nAssociation Rules:")
rules
```

Provide explanation about __antecedent support__, __consequent support__, __support__, __confidence__, __lift__, __leverage__ and __conviction__

```python
Association rule mining involves discovering patterns in data where certain events or conditions are associated with others. Several metrics aid in evaluating these rules. Antecedent support measures how often the antecedent occurs, while consequent support assesses the frequency of the consequent. Support quantifies the joint occurrence of both antecedent and consequent, highlighting strong associations. Confidence indicates the likelihood of finding the consequent given the antecedent, with higher values signifying stronger rules. Lift compares the observed support to the expected support under independence, 
and values greater than 1 suggest a meaningful association. Leverage assesses how much more antecedent and consequent occur together than expected by chance. 
Conviction measures the dependence of the consequent on the antecedent, with higher values indicating stronger dependence. 
These metrics collectively provide insights into the significance and reliability of association rules, aiding decision-making in various domains.
```
